# 天気アイコンの作成

まず，cloud.png, fog.png, rain.png, snow.png, sun.png, thunderstorm.png, wind.pngをダウンロードして./imgs以下に配置してから  
以下のコードを実行する

天気コードはhttps://weather.yukigesho.com/code.htmlを参考にした

```python
import numpy as np
from PIL import Image, ImageDraw
```

```python
# 設定
img_dir = "./"
```

```python
# 単体画像の作成
def create_single_image(img_dir, img_name, output_name):
    # 画像の読み込み
    new_image = Image.open(img_dir + img_name)
    new_image.save(img_dir + output_name)
    # new_image.show()

# img1のちimg2の画像の作成
def create_then_image(img_dir, img1_name, img2_name, output_name):
    # 画像の読み込み
    img1 = Image.open(img_dir + img1_name)
    img2 = Image.open(img_dir + img2_name)
    # img2の倍率
    x = 0.6
    img2_size = (int(img2.width * x), int(img2.height * x))
    img2 = img2.resize(img2_size)
    # 新しい画像を作成
    new_size = (int(img1.size[0]*(1+x)), int(img1.size[1]*1.2))
    new_image = Image.new('RGBA', new_size, (255, 255, 255, 0))
    new_image.paste(img1, (0, 0))
    new_image.paste(img2, (new_image.width - img2.width, img1.height - img2.height))
    # ImageDrawオブジェクトを作成
    draw = ImageDraw.Draw(new_image)
    # ポイントの座標を設定
    arrow_points = [
        # (x, y)
        (int(new_size[0]*0.3), int((new_size[1]-(new_size[1]-img1.size[1])*0.4))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img1.size[1])*0.4))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img1.size[1])*0.1))),
        (int(new_size[0]*0.7), int((new_size[1]-(new_size[1]-img1.size[1])*0.5))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img1.size[1])*0.9))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img1.size[1])*0.6))),
        (int(new_size[0]*0.3), int((new_size[1]-(new_size[1]-img1.size[1])*0.6))),
    ]
    # 白い矢印を描画
    draw.polygon(arrow_points, fill='white')
    # 結果の画像を保存
    new_image.save(img_dir + output_name)
    # new_image.show()

# img1のちimg2またはimg3の画像の作成
def create_then_image2(img_dir, img1_name, img2_name, img3_name, output_name):
    # 画像の読み込み
    img1 = Image.open(img_dir + img1_name)
    img2 = Image.open(img_dir + img2_name)
    img3 = Image.open(img_dir + img3_name)
    # img2, img3の倍率
    x = 0.5
    img2_size = (int(img2.width * x), int(img2.height * x))
    img2 = img2.resize(img2_size)
    img3_size = (int(img3.width * x), int(img3.height * x))
    img3 = img3.resize(img3_size)
    # 新しい画像を作成
    new_size = (int(img1.size[0]*(1+x)), int(img1.size[1]*1.2))
    new_image = Image.new('RGBA', new_size, (255, 255, 255, 0))
    new_image.paste(img1, (0, 0))
    new_image.paste(img2, (new_image.width - img2.width, 0))
    new_image.paste(img3, (new_image.width - img3.width, img1.height - img3.height))
    # ImageDrawオブジェクトを作成
    draw = ImageDraw.Draw(new_image)
    # ポイントの座標を設定
    arrow_points = [
        # (x, y)
        (int(new_size[0]*0.3), int((new_size[1]-(new_size[1]-img1.size[1])*0.4))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img1.size[1])*0.4))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img1.size[1])*0.1))),
        (int(new_size[0]*0.7), int((new_size[1]-(new_size[1]-img1.size[1])*0.5))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img1.size[1])*0.9))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img1.size[1])*0.6))),
        (int(new_size[0]*0.3), int((new_size[1]-(new_size[1]-img1.size[1])*0.6))),
    ]
    # 白い矢印を描画
    draw.polygon(arrow_points, fill='white')
    # 結果の画像を保存
    new_image.save(img_dir + output_name)
    # new_image.show()

# img1またはimg2のちimg3の画像の作成
def create_then_image3(img_dir, img1_name, img2_name, img3_name, output_name):
    # 画像の読み込み
    img1 = Image.open(img_dir + img1_name)
    img2 = Image.open(img_dir + img2_name)
    img3 = Image.open(img_dir + img3_name)
    # img1, img2の倍率
    x = 0.5
    img1_size = (int(img1.width * x), int(img1.height * x))
    img1 = img1.resize(img1_size)
    img2_size = (int(img2.width * x), int(img2.height * x))
    img2 = img2.resize(img2_size)
    # 新しい画像を作成
    new_size = (int(img3.size[0]*(1+x)), int(img3.size[1]*1.2))
    new_image = Image.new('RGBA', new_size, (255, 255, 255, 0))
    new_image.paste(img1, (0, 0))
    new_image.paste(img2, (0, img3.height - img2.height))
    new_image.paste(img3, (new_image.width - img3.width, 0))
    # ImageDrawオブジェクトを作成
    draw = ImageDraw.Draw(new_image)
    # ポイントの座標を設定
    arrow_points = [
        # (x, y)
        (int(new_size[0]*0.3), int((new_size[1]-(new_size[1]-img3.size[1])*0.4))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img3.size[1])*0.4))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img3.size[1])*0.1))),
        (int(new_size[0]*0.7), int((new_size[1]-(new_size[1]-img3.size[1])*0.5))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img3.size[1])*0.9))),
        (int(new_size[0]*0.6), int((new_size[1]-(new_size[1]-img3.size[1])*0.6))),
        (int(new_size[0]*0.3), int((new_size[1]-(new_size[1]-img3.size[1])*0.6))),
    ]
    # 白い矢印を描画
    draw.polygon(arrow_points, fill='white')
    # 結果の画像を保存
    new_image.save(img_dir + output_name)
    # new_image.show()

# img1時々img2の画像の作成
def create_sometimes_image(img_dir, img1_name, img2_name, output_name):
    # 画像の読み込み
    img1 = Image.open(img_dir + img1_name)
    img2 = Image.open(img_dir + img2_name)
    # img2の倍率
    x = 0.6
    img2_size = (int(img2.width * x), int(img2.height * x))
    img2 = img2.resize(img2_size)
    # 新しい画像を作成
    new_size = (int(img1.size[0]*(1+x)), int(img1.size[1]))
    new_image = Image.new('RGBA', new_size, (255, 255, 255, 0))
    new_image.paste(img1, (0, 0))
    new_image.paste(img2, (new_image.width - img2.width, img1.height - img2.height))
    # 結果の画像を保存
    new_image.save(img_dir + output_name)
    # new_image.show()

# img1時々img2またはimg3画像の作成
def create_sometimes_image2(img_dir, img1_name, img2_name, img3_name, output_name):
    # 画像の読み込み
    img1 = Image.open(img_dir + img1_name)
    img2 = Image.open(img_dir + img2_name)
    img3 = Image.open(img_dir + img3_name)
    # img2, img3の倍率
    x = 0.5
    img2_size = (int(img2.width * x), int(img2.height * x))
    img2 = img2.resize(img2_size)
    img3_size = (int(img3.width * x), int(img3.height * x))
    img3 = img3.resize(img3_size)
    # 新しい画像を作成
    new_size = (int(img1.size[0]*(1+x)), int(img1.size[1]))
    new_image = Image.new('RGBA', new_size, (255, 255, 255, 0))
    new_image.paste(img1, (0, 0))
    new_image.paste(img2, (new_image.width - img2.width, 0))
    new_image.paste(img3, (new_image.width - img3.width, img1.height - img3.height))
    # 結果の画像を保存
    new_image.save(img_dir + output_name)
    # new_image.show()

# img1またはimg2時々img3の画像の作成
def create_sometimes_image3(img_dir, img1_name, img2_name, img3_name, output_name):
    # 画像の読み込み
    img1 = Image.open(img_dir + img1_name)
    img2 = Image.open(img_dir + img2_name)
    img3 = Image.open(img_dir + img3_name)
    # img1, img2の倍率
    x = 0.5
    img1_size = (int(img1.width * x), int(img1.height * x))
    img1 = img1.resize(img1_size)
    img2_size = (int(img2.width * x), int(img2.height * x))
    img2 = img2.resize(img2_size)
    # 新しい画像を作成
    new_size = (int(img3.size[0]*(1+x)), int(img3.size[1]))
    new_image = Image.new('RGBA', new_size, (255, 255, 255, 0))
    new_image.paste(img1, (0, 0))
    new_image.paste(img2, (0, img3.height - img2.height))
    new_image.paste(img3, (new_image.width - img3.width, 0))
    # 結果の画像を保存
    new_image.save(img_dir + output_name)
    # new_image.show()
```
